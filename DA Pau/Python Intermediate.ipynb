```python
offset = 8
while offset != 0:
    if offset > 0:
        print('correcting...')
        offset -= 1
        print(f'offset value = {offset}')
    elif offset < 0:
        print('correcting...')
        offset += 1
        print(f'offset value = {offset}')
        
```
